```python
import matplotlib.pyplot as plt
```

```python
x = ['Cold As You (2006)','White Horse (2008)','Sparks Fly (2010)','I Almost Do (2012)','Welcome To New York (2014)','Ready For It (2017)','London Boy (2019)','Peace (mid-2020)','Champagne Problems (late-2020)','Labyrinth (2022)']
y = [.2204,.2175,.2017,.1413,.1391,.1396,.1397,.2334,.4090,.1254]
```

```python

plt.tight_layout()
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(6)
plt.xlabel('Song Title')
plt.ylabel('Proportion Unique Words')
plt.title('Taylor Swift Lyrical Complexion')
plt.plot(x, y)
```

```python
data = {'Cool':.0827, '1000 Doves':.1764, 'Unholy':.2017,
        'Blinding Lights':.1760}
courses = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))
```

```python
plt.bar(courses, values, color ='blue',
        width = 0.4)
 
plt.xlabel("Song Title")
plt.ylabel("Proportion of Unique Words")
plt.title("Students enrolled in different courses")
plt.show()
```
